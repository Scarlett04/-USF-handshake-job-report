```python
import pandas as pd
import numpy as np
import math

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("csv/modified_data.csv")
```

```python
df.head(10)
```

```python
df.columns
```

```python
# Create a list of all company locations and put them in a new column
df['Location List'] = df['Company Location'].apply(lambda x: x.split('\n') if '\n' in x else x)
df.head(10)
```

```python
# If 'Company Location' row contains multiple locations, create a new row for each location
df = df.explode('Location List')
df.head(10)
```

```python
# Create a new column 'States abbreviation' and extract the state abbreviation from 'Location List' column
df['States abbreviation'] = df['Location List'].str.findall(r'\b[A-Z]{2}\b').str[-1]
df.head(10)
```

```python
df.drop(df[df['States abbreviation'].isnull()].index, inplace=True)
```

```python
location = df[['ID', 'States abbreviation']]
location
```

```python
location.to_csv('csv/location.csv', index=False)
```
